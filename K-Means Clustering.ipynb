```python
from pybaseball import batting_stats
from sklearn import preprocessing
import pandas
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load dataset
initial_data = batting_stats(2019,qual=400)
data = initial_data[['OBP','Age','O-Swing%','Z-Swing%']]
data = data.dropna()
# data['O-Swing%'].isnull().sum()

data_standardized = preprocessing.scale(data)
data_standardized = pandas.DataFrame(data)

# use to find number of clusters I should start with
# plt.figure(figsize=(10, 8))
# wcss = []
# for i in range(1, 11):
#     kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
#     kmeans.fit(data_standardized)
#     wcss.append(kmeans.inertia_)
# plt.plot(range(1, 11), wcss)
# plt.title('The Elbow Method')
# plt.xlabel('Number of clusters')
# plt.ylabel('WCSS')
# plt.show()

kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(data_standardized)
y_kmeans1=y_kmeans
y_kmeans1=y_kmeans+1
cluster = pandas.DataFrame(y_kmeans1)
data['cluster'] = cluster
kmeans_mean_cluster = pandas.DataFrame(round(data.groupby('cluster').mean(),2))

joined = pandas.merge(initial_data,data,left_index=True,right_index=True)
joined = pandas.DataFrame(joined)

#Summary variables below
cluster_averages = joined.groupby('cluster').describe()
cluster_averages.T

cluster_counts = joined.groupby('cluster')['cluster'].value_counts()
best_obp_cluster = joined.loc[joined.groupby('cluster')['OBP_x'].idxmax()]
best_obp_cluster
```
