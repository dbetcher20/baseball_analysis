```python
#Comparing players over the course of 4 seasons with at least 1000 at bats to determine differences in pitches seen
#by elite hitters and below average hitters

from pybaseball import batting_stats, playerid_lookup
import pandas
import numpy

good_hitters = batting_stats(start_season = 2016, end_season = 2019, qual = 1000, ind = 0).sort_values('wRC+', ascending = False).head(50)
bad_hitters = batting_stats(start_season = 2016, end_season = 2019, qual = 1000, ind = 0).sort_values('wRC+', ascending = False).tail(50)
#make a category for all other players to compare all three

# for col in good_hitters.columns:
#     print(col)

gh_averages = good_hitters.agg({'Zone%': ['mean'], 'F-Strike%': ['mean'], 'FC-Z (pfx)': ['mean'], 'FC-X (pfx)': ['mean'], 'FT-Z (pfx)': ['mean'], 'FA-Z (pfx)': ['mean'], 'FT-X (pfx)': ['mean'], 'FA-X (pfx)': ['mean'], 'CH%': ['mean'], 'CHv': ['mean'], 'FBv': ['mean'], 'FB% (Pitch)': ['mean'], 'WAR': ['mean'], 'wRC+': ['mean'], 'OBP': ['mean'], 'BB%': ['mean'], 'AVG': ['mean'], 'Pitches': ['mean'], 'Balls': ['mean'], 'Strikes': ['mean']})
gh_averages['Hitter_Type'] = 'Good'

bh_averages = bad_hitters.agg({'Zone%': ['mean'], 'F-Strike%': ['mean'], 'FC-Z (pfx)': ['mean'], 'FC-X (pfx)': ['mean'], 'FT-Z (pfx)': ['mean'], 'FA-Z (pfx)': ['mean'], 'FT-X (pfx)': ['mean'], 'FA-X (pfx)': ['mean'], 'CH%': ['mean'], 'CHv': ['mean'], 'FBv': ['mean'], 'FB% (Pitch)': ['mean'], 'WAR': ['mean'], 'wRC+': ['mean'], 'OBP': ['mean'], 'BB%': ['mean'], 'AVG': ['mean'], 'Pitches': ['mean'], 'Balls': ['mean'], 'Strikes': ['mean']})
bh_averages['Hitter_Type'] = 'Bad'
#look into ways to help correct for lefties/righties pitch movement numbers, can I do absolute value?
final = gh_averages.append(bh_averages)

final['K_Pitch_%'] = final['Strikes']/final['Pitches']
final.drop(columns = ['Pitches','Strikes','Balls'], inplace = True)

good = final[final['Hitter_Type'] == 'Good']
bad = final[final['Hitter_Type'] == 'Bad']

final
```
